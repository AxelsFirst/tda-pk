```python
import os
import sys
import random

from string import ascii_uppercase as letters
```

```python
sys.path.append(os.path.join(os.getcwd(), '..'))
```

```python
from tda.metrics import euclidean_metric
from tda.point import Point
from tda.vr_complex import VietorisRipsComplex
from tda.plots import plot_graph, plot_faces, plot_simplices
```

```python
pts = list()
for i in range(len(letters)):
    coords = [random.random() for _ in range(2)]
    pts.append(Point(name=letters[i], coords=coords))

vrc = VietorisRipsComplex(
    points=pts,
    epsilon=1/5,
    metric=euclidean_metric
)

vrc.create_graph()
```

```python
plot_graph(
    complex=vrc,
    save=True, 
    file_name='images/example_plot_graph'
)
```

```python
plot_graph(
    complex=vrc,
    save=True,
    draw_ball=True,
    file_name='images/example_plot_graph_with_balls'
)
```

```python
pts = list()
for i in range(len(letters)):
    coords = [random.random() for _ in range(2)]
    pts.append(Point(name=letters[i], coords=coords))

vrc = VietorisRipsComplex(
    points=pts,
    epsilon=1/4,
    metric=euclidean_metric
)

vrc.create_graph()
s = vrc.find_simplices()
f = vrc.find_faces()
```

```python
plot_faces(
    complex=vrc,
    face_dimension=2,
    save=True,
    file_name='images/example_plot_faces'
)
```
